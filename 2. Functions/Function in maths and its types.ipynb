There are 5 main types of basic functions: power functions, exponential functions, logarithms, trigonometric functions and inverse trigonometric functions, but for our course it is necessary to know only two first types.

First type - power functions which can be written in a form $$\mathbf{f(x)=cx^p}$$ You can play with different values of c and p in a following code:

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt

@interact(p=(-5,5,0.5),c=(-2,2,0.5))
def f(p=0,c=1):    
    if (p==int(p)):
        if (p>=0 or c==0):             
            x=np.arange(-1.995,2,0.01)
            y=c*x**p
            plt.plot(x,y,'-b') 
        else:
            ax=abs(30/c)**(1/p)            
            x1=np.arange(-2,-ax,0.01)
            x2=np.arange(ax,2,0.01)
            y1=c*x1**p
            y2=c*x2**p
            plt.plot(x1,y1,'-b') 
            plt.plot(x2,y2,'-b')
        
    else:
        if (p>=0 or c==0):             
            x=np.arange(0.005,2,0.01)
            y=c*x**p
            plt.plot(x,y,'-b') 
        else:
            ax=abs(30/c)**(1/p)            
            x=np.arange(ax,2,0.01)            
            y=c*x**p            
            plt.plot(x,y,'-b') 
            
    s="$f(x)="+str(c)+"x^{"+str(p)+"}$"
    plt.title(s)
    plt.show()
```

Second type - exponential functions which can be written in a form $$\mathbf{f(x)=ca^x}$$ where $a>0,a\neq 1$. Number a is called a base of an exponent. In the following demo you can change parametres c and a as you want:

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt

@interact(a=(0.1,3,0.3),c=(-2,2,0.5))
def f(a=2,c=1):    
    x=np.arange(-2,2,0.01)
    y=c*np.exp(np.log(a)*x)
    plt.plot(x,y,'-b')
            
    s="$f(x)="+str(c)+str(a)+"^x$"
    plt.title(s)
    plt.show()
```

Regarding exponential function, there exists special number $e=2.71828...$, which is sometimes called Euler's number. It is quite often used as a base of an exponent, in other words - number a, because expressions with this number has several benefits, for example it is much easier to do mathematical operations from calculus, which will be partly reviewed in the second part of this lecture, with these expressions.

Later on, we will use this functions and also some functions that can be get by summing, multiplying, subtracting, dividing and raising to the power basic functions of the first and second type. For example $$f(x)=\frac{e^x-e^{-x}}{e^x+e^{-x}}, g(x)=\frac{1}{1+x^2}$$

A lot of other functions exist but they are not very common in machine learning. That is why, we will learn about only two more functions which will be used as a replacement of different "spiky" functions. These two functions are $$f(x)=b\cos(ax+\phi)$$ and $$f(x)=b\sin(ax+\phi)$$

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt

@interact(a=(0.1,3,0.3),b=(-2,2,0.5),ph=(0,4,0.5))
def f(a=2,b=1,ph=0):    
    x=np.arange(-5,5,0.01)
    y=b*np.cos(a*x+ph)
    plt.plot(x,y,'-b')
            
    s="$f(x)="+str(b)+"cos("+str(a)+"x+"+str(ph)+")$"
    plt.title(s)
    plt.show()
```

## Function of several variables

Function of several inputs takes as input n real numbers, commonly represented by the variables $x1, x2, ..., x_n$, for producing another real number, the value of the function, commonly denoted $f(x1, x2, ..., x_n)$.

### Examples

#### Function of two variables

Altitude of curved surface above sea level depending on 2 coordinates
$$ H(x,y)=\sin{4x}+\cos{3y}-x^2 $$
A function of this type can be represented using contour plot - the same method used in cartography:

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

matplotlib.rcParams['xtick.direction'] = 'out'
matplotlib.rcParams['ytick.direction'] = 'out'

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z=np.sin(4*X)+np.cos(3*Y)-X**2

plt.figure(figsize=(10,7))
CS = plt.contour(X, Y, Z,levels=np.arange(-8,1,1))
plt.clabel(CS, inline=1, fontsize=10)
plt.title('Contour plot')
plt.show()
```

Other method is drawing 3D graph of the function $z=f(x,y)$, where each point on OXY plane correspond to certain value of z. The following code builds 3D graph of the same altitude function $H(x,y)$

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure(figsize=(10,7))
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-3, 3, 0.25)
Y = np.arange(-2, 2, 0.25)
X, Y = np.meshgrid(X, Y)
Z=np.sin(4*X)+np.cos(3*Y)-X**2

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

#### Function with three variables

Functions of this type are very commom in Physics, because a lot of scalar values depend on 3 coordinates in space: $x,y,z$. For example, in stationary conditions (when values doesn't depend on time) temperature, density, saltiness of water can depend in the general case on 3 coordinates. Despite its commonness, it is quite hard to plot exact graph of such functions. Obviously, we cannot build it in 4 dimensions (we build graph of 1 variable's function in 2D and graph of 2 variables' function in 3D, and by analogy...), because our brains are not able to perceive it. That is why, we can use methods of introducing new dimension as plotting each point in 3D in different color, which will represent value of function or plotting points of different size, which is called bubble chart.

This is example of bubble chart for function $T(x,y,z)=\cos(x^2+y^2)e^{-z}$

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure(figsize=(10,7))
ax = fig.gca(projection='3d')

# Make data.
step=0.4
X = np.arange(-0.8, 0.8, step)
Y = np.arange(-0.8, 0.8, step)
Z = np.arange(-0.8, 0.8, step)
X, Y, Z = np.meshgrid(X, Y, Z)


# Plot the surface.
ax.scatter(X, Y, Z, s=500*np.cos(X**2+Y**2)*np.exp(-Z),color='r')

plt.show()
```

To be honest it is pretty confusing, let's try color plot of the same function:

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.colors as colors
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure(figsize=(10,7))
ax = fig.gca(projection='3d')

# Make data.
step=0.4
X = np.arange(-0.8, 0.8, step)
Y = np.arange(-0.8, 0.8, step)
Z = np.arange(-0.8, 0.8, step)
X, Y, Z = np.meshgrid(X, Y, Z)

cNorm  = colors.Normalize(vmin=0, vmax=1)
scalarMap = cm.ScalarMappable(norm=cNorm, cmap=cm.coolwarm)

# Plot the surface.
ax.scatter(X, Y, Z, s=144,c=scalarMap.to_rgba(np.cos(X**2+Y**2)*np.exp(-Z)))

plt.show()
```

We can see that temperature is higher the lower and closer to center point is. Still, this methods are not very effective and more important *not precise enough*. If we could somehow evaluate or at least distinguish values corresponding to different points at 3D graph, it is just almost impossible to make with graph of 3 variables' function. <br>
